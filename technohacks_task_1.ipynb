## **HANDLING OUTLIERS**

```python
from scipy import stats
```

```python
#dentifying Outliers
z_scores_age = np.abs(stats.zscore(a['Age']))
z_scores_fare = np.abs(stats.zscore(a['Fare']))
threshold = 3
outliers_age = z_scores_age > threshold
outliers_fare = z_scores_fare > threshold
```

```python
#removing outliers
df = a[outliers_age]
df = a[outliers_fare]
```
